```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Read in the Total Playlist CSV

```python
#read the sound_salvation_fillin.csv file
total_playlists = pd.read_csv('sound_salvation_fillin_playlists.csv')
```

```python
total_playlists
```

Read in the Date and DJ CSV

```python
dd_df = pd.read_csv('date_and_dj_df.csv')
```

```python
dd_df
```

```python
#drop the unnamed column in dd_df
dd_df = dd_df.drop(columns=['Unnamed: 0'])
```

```python
dd_df
```

Merge the two dataframes together on the URL 

```python
#merge the two dataframes and fill forward the dj and date columns
merged_df = pd.merge(total_playlists, dd_df, how='left', on='url')
```

```python
merged_df
```

```python
#drop the unnamed column in merged_df
merged_df = merged_df.drop(columns=['Unnamed: 0'])
#index the merged_df by the date column
merged_df = merged_df.set_index('date')

merged_df
```

Order the merged dataframe by date and drop the URL column 

```python
#sort the index
merged_df = merged_df.sort_index()

#change the album year column to an integer
# merged_df['album_year'] = merged_df['album_year'].astype(int)
#can't do because there are NaN values

#drop the URL column
merged_df = merged_df.drop(columns=['url'])
#print entire note column to the screen
pd.set_option('display.max_colwidth', -1)



merged_df
```

```python
merged_df.head(20)
```

```python
#drop duplicate rows
merged_df = merged_df.drop_duplicates()
```

```python
merged_df
```

```python
#find missing value counts
merged_df.isnull().sum()
```

```python
#drop any row with anyone  other than steve pick in the dj column
merged_df = merged_df[merged_df['dj'] == 'Steve Pick']
merged_df
```

```python
#drop the nan values in the album_year column
merged_df = merged_df.dropna(subset=['album_year'])
#convert the album_year column to an integer
merged_df['album_year'] = merged_df['album_year'].astype(int)
merged_df
```

```python
#save the merged_df to a csv file
merged_df.to_csv('Sound_Salvation_Final.csv')
```

Sort the time column to times between 7:00 and 7:05 to find the most common artists that lead off the show!

```python
#find the top artists played at 7:00 to 7:05 AM (the first 5 minutes of the show)
plt.figure(figsize=(20,10))
first_plays = merged_df[merged_df['time'] <= '7:05 AM'].sort_values(by='date', ascending=False)
#plot the top 10 artists with seaborn
sns.countplot(y="artist", data=first_plays, order=first_plays['artist'].value_counts().iloc[:10].index, palette="Set2")
```

Sort the time column to times between 8:00 and 8:05 to find the most common artists that lead off the second hour!

```python
plt.figure(figsize=(20,10))
#find the rows where the time is greater that 8:00 AM but less than 8:05 AM
secondhour_firstplays = merged_df[(merged_df['time'] >= '8:00 AM') & (merged_df['time'] <= '8:05 AM')].sort_values(by='date', ascending=False)
#plot the top 10 artists with seaborn
sns.countplot(y="artist", data=secondhour_firstplays, order=secondhour_firstplays['artist'].value_counts().iloc[:10].index, palette="Set2")

# secondhour_firstplays['artist'].value_counts().head(10).plot(kind='bar', figsize=(20,10))
```

Sort the time column to times between 9:00 and 9:05 to find the most common artists that kick off the final hour!

```python
plt.figure(figsize=(20,10))
#find the rows where the time is greater that 9:00 AM but less than 9:05 AM
thirdhour_firstplays = merged_df[(merged_df['time'] >= '9:00 AM') & (merged_df['time'] <= '9:05 AM')].sort_values(by='date', ascending=False)
#plot the top 10 artists with seaborn
sns.countplot(y="artist", data=thirdhour_firstplays, order=thirdhour_firstplays['artist'].value_counts().iloc[:10].index, palette="Set2")
```

Find out the artists that close out the show

```python
plt.figure(figsize=(20,10))
#find the rows where the time is greater that 9:00 AM but less than 9:05 AM
last_plays = merged_df[(merged_df['time'] >= '9:50 AM') & (merged_df['time'] <= '9:59 AM')].sort_values(by='date', ascending=False)
#plot the top 10 artists with seaborn
sns.countplot(y="artist", data=last_plays, order=last_plays['artist'].value_counts().iloc[:10].index, palette="Set2")
```
