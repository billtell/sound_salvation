```python
def sound_salvation_dateDJ_spider(url):

    """Perform web scraping for  sound salvation playlist given the available link"""

    r = requests.get(url)
    rock_soup = BeautifulSoup(r.text, 'html')
    
    #get the date and dj
    # days = []
    # months = []
    # years = []
    # djs = []
    links = []

    month = rock_soup.find_all('span', attrs={'class':'month'})
    day = rock_soup.find_all('span', attrs={'class':'day'})
    year = rock_soup.find_all('span', attrs={'class':'year'})
    dj = rock_soup.find_all('div', attrs={'class':'dj'})

    # append the data to the empty lists
    months = ([m.string for m in month])
    days = ([d.string for d in day])
    years = ([y.string for y in year])
    djs = ([d.string for d in dj])

    # append the data to the empty lists
    links.append(url)
    # months.append(month)
    # days.append(day)
    # years.append(year)
    # djs.append(dj)
    
    my_date_dj_dict = {
    'month': months, 
    'day': days, 
    'year': years, 
    'dj': djs, 
    'link': links}

    df = pd.DataFrame(my_date_dj_dict)
    #fill in the missing values
    df['month'] = df['month'].fillna(method='ffill')
    df['day'] = df['day'].fillna(method='ffill')
    df['year'] = df['year'].fillna(method='ffill')
    df['dj'] = df['dj'].fillna(method='ffill')
    df['link'] = df['link'].fillna(method='ffill')

    return df
    
```
